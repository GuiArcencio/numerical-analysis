```python
import numpy as np
```

```python
def lagrange_interpolation(X=None, Y=None, P=None):
    if P != None:
        X = list()
        Y = list()
        for x,y in P:
            X.append(x)
            Y.append(y)

    def lagr_polynomial(x):
        result = 0.
        for j, xj in enumerate(X):
            product = 1.
            quotient = 1.
            for m, xm in enumerate(X):
                if (m != j):
                    product *= x - xm
                    quotient *= xj - xm
            result += Y[j] * (product / quotient)
        return result
    return lagr_polynomial
```

```python
X = np.linspace(0, 2*np.pi, 6)
Y = np.sin(X)

poly = lagrange_interpolation(X,Y)

x_points = np.linspace(min(X), max(X), 1000)
y_points = poly(x_points)

import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
plt.style.use('ggplot')
plt.scatter(X, Y, c='blue')
plt.plot(x_points, y_points, 'r-')

plt.show()
```
