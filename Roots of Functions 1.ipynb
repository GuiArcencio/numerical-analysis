```python
import numpy as np
```

```python
def bisection(f, a, b, n=10):
    left = a
    right = b
    x = (right + left)/2
    for _ in range(n):
        if f(x)*f(left) < 0:
            right = x
            x = (right + left)/2
        else:
            left = x
            x = (right + left)/2
    return x
```

```python
bisection(lambda x: np.sqrt(x) - np.cos(x), 0, 1)
```
