```python
import numpy as np
```

```python
def newton_rhapson(f, x0, f_prime=None, N=10):
    epsilon = 1e-6
    if f_prime == None:
        # Approximate deriavtive
        f_prime = lambda x: (f(x+epsilon) - f(x-epsilon))/(2*epsilon)
    x = x0
    for _ in range(N):
        x = x - f(x)/f_prime(x)

    return x

def secant_method(f, x0, x1, N=10):
    if N == 0: return x0
    if N == 1: return x1
    x = (x0 * f(x1) - x1 * f(x0))/(f(x1) - f(x0))
    for _ in range(N-2):
        x0 = x1
        x1 = x
        if f(x1) - f(x0) == 0:
            break
        x = (x0 * f(x1) - x1 * f(x0))/(f(x1) - f(x0))

    return x
```

```python
p = list()
for i in range(6):
    p.append(newton_rhapson(lambda x: x*np.sin(x), 0.9, lambda x: np.sin(x) + x*np.cos(x), i))
```
