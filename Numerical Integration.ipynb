```python
import numpy as np
```

```python
def trapezoid(f, a, b, n):
    x, h = np.linspace(a, b, n+1, retstep=True)
    for xi, fxi in zip(x, f(x)):
        print(f'{xi}\t{fxi}')
    total = 0.
    for i, xi in enumerate(x):
        total += (1 if i == 0 or i == x.shape[0] - 1 else 2) * f(xi)
    return (h/2) * total

def simpson(f, a, b, n):
    x, h = np.linspace(a, b, n+1, retstep=True)
    total = 0.
    for i, xi in enumerate(x):
        coefficient = 0.
        if i == 0 or i == x.shape[0] - 1: coefficient = 1
        elif i % 2 == 0: coefficient = 2
        else: coefficient = 4

        total += coefficient * f(xi)
    return (h/3) * total
```

```python
p = lambda x: 5.7384 - 7.7245*x - 9.6988 * x**2 - 6.5542 * x**3 + 8.7136 * x**4 - 9.044 * x**5
simpson(p, -1.1228, 1.859, 4)
```

```python
f = lambda x: np.exp(x)

trapezoid(f, 0, 9, 12)
```

```python
f = lambda x: x*x*np.sqrt(x*x + 1)

trapezoid(f, 1, 2, 7)
```
