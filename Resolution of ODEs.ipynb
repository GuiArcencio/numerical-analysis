```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler(f, t0, y0, n, h=0.1):
    t = [t0]
    y = [y0]
    for i in range(n):
        t.append(t[i] + h)
        y.append(y[i] + h*f(t[i], y[i]))
    return t, y
```

```python
x, y = euler(lambda t,y: y + np.sin(t), 0, 0.5, 11)
plt.style.use('ggplot')

plt.plot(x,y,'r-')

plt.show()
```
